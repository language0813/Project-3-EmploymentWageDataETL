```python
# Import the dependency
import pandas as pd
```

# Step 1: Extract


```python
# Load the data into a Pandas DataFrame
file_path = 'Data/all_data_M_2023.xlsx'
all_owes_df = pd.read_excel(file_path, sheet_name='All May 2023 data')
all_owes_df.head()
```

# Step 2: Transform the data

```python
# Narrow down the data to State level
state_oews_df = all_owes_df.loc[all_owes_df['AREA_TYPE'] == 2, :]
state_oews_df.head()
```

```python
# Select the columns we need
columns_to_keep = [
    'AREA', 
    'AREA_TITLE', 
    'PRIM_STATE',
    'OCC_CODE', 
    'OCC_TITLE', 
    'O_GROUP',
    'TOT_EMP',
    'EMP_PRSE',
    'JOBS_1000',
    'LOC_QUOTIENT',
    'H_MEAN',
    'A_MEAN',
    'MEAN_PRSE',
    'H_PCT25',
    'H_MEDIAN',
    'H_PCT75',
    'A_PCT25',
    'A_MEDIAN',
    'A_PCT75'
]

cleaned_df = state_oews_df[columns_to_keep]
cleaned_df.head()
```

```python
# Rename the columns for preparing to load data into database
renamed_oews_df = cleaned_df.rename(columns={
    'AREA': 'area',
    'AREA_TITLE': 'area_title',
    'PRIM_STATE': 'prim_state',
    'OCC_CODE': 'occ_code',
    'OCC_TITLE': 'occ_title',
    'O_GROUP': 'o_group',
    'TOT_EMP': 'tot_emp',
    'EMP_PRSE': 'emp_prse',
    'JOBS_1000': 'jobs_1000',
    'LOC_QUOTIENT': 'loc_quotient',
    'H_MEAN': 'h_mean',
    'A_MEAN': 'a_mean',
    'MEAN_PRSE': 'mean_prse',
    'H_PCT25': 'h_pct25',
    'H_MEDIAN': 'h_median',
    'H_PCT75': 'h_pct75',
    'A_PCT25': 'a_pct25',
    'A_MEDIAN': 'a_median',
    'A_PCT75': 'a_pct75'
})
renamed_oews_df.head()
```

```python
# Check the data type before loading them into database
renamed_oews_df.info()
```

```python
# The list of columns to convert the data type
columns_to_convert = [
    'tot_emp',
    'emp_prse',
    'jobs_1000',
    'loc_quotient',
    'h_mean',
    'a_mean',
    'mean_prse',
    'h_pct25',
    'h_median',
    'h_pct75',
    'a_pct25',
    'a_median',
    'a_pct75'
]
```

```python
# Convert the specified columns to numeric, coercing errors
renamed_oews_df[columns_to_convert] = renamed_oews_df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

renamed_oews_df.info()
```

```python
# Drop the data that are not available
final_df = renamed_oews_df.dropna(how='any')
final_df.head()
```

```python
# Check the result after dropping the data
final_df.info()
```

# Step 3: Load the Data

```python
# Export the data to CSV
final_df.to_csv('data/OEWSData.csv', index=False)
```
